```python
k = 1.10886383
a = 1
A2 = (a + (np.cosh(k*a))**2 / k  + (np.sinh(2*k*a)) / 2*k )

A = np.sqrt(A2)

x1 = np.linspace(-15 , -1 , 1000)
x2 = np.linspace(-1 , 1 , 1000)
x3 = np.linspace(1 , 15 , 1000)

y1 = A*np.cosh(k*a)*np.exp(k*(x1+a))
y2 = A*np.cosh(k*x2)
y3 = A*np.cosh(k*a)*np.exp(-k*(x3-a))
```

```python
plt.figure(figsize=(18,11))

plt.plot(x1 , y1 , color = 'darkviolet' , label = '$\psi(x)$ $x<-a$')
plt.plot(x2 , y2 , color = 'crimson' ,label = '$\psi(x)$ $-a<x<a$' )
plt.plot(x3 , y3 , color = 'springgreen' , label = '$\psi(x)$ $x>a$')

plt.title('Función de onda en el espacio' , fontsize = 25 )
plt.xlabel('$x$ $(a.u)$' , fontsize = 25 , color = 'black')
plt.ylabel('$\psi(x)$ ' , fontsize = 25 , color = 'black')

plt.rc('lines', linewidth=2.3 , color='r')


plt.xlim(-5 , 5)

plt.legend(loc = "upper left" , prop={'size':23})

plt.savefig('1-71-funciononda-app.png')
```
