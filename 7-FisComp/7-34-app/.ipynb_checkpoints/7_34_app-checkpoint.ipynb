```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Apartado a

```python
col = ['raiz' , 'd']

datos = pd.read_table('dats-a-b-c/7-34-a-app.dat', sep= "\s+" , names = col , header = None)
datos
```

```python
plt.figure(figsize = (16,9))

plt.plot(datos.raiz , datos.d )
```

```python
x = np.array(datos['raiz']).reshape((-1,1))
y = np.array(datos['d'])
```

```python
model = LinearRegression()

model.fit(x , y)

R_sq = model.score(x, y)

m = model.coef_            #Pendiente

ord = model.intercept_     #Ordenada en el origen

r2 = R_sq                  #CorrelaciÃ³n

print(m) , print(ord) , print(r2)
```

```python
a = np.linspace(0 , 317 , 5000)

b = 1.2*a  + ord

plt.figure(figsize = (16,9))


plt.plot(datos.raiz , datos.d )
plt.plot(a , b, color = 'r')
```
