```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Apartado a

```python
col = ['i' , 'd']
datos = pd.read_table('7-34-a-app.dat', sep= "\s+" , names = col , header = None)

datos
```

```python
plt.style.use('ggplot')

a = np.linspace(0 , 32 , 5000)

b = a  

plt.figure(figsize = (16,9))

plt.plot(np.sqrt(datos.i), datos.d , color = 'b' , linewidth = '1' , label= '$d_{rms}$' )
plt.plot(a , b, color = 'r' , linewidth = '1.9' , label = 'recta $y=x$' )

plt.title('Distancia $d_{rms}$ al hotel calles de lado $L=$1 m' , fontsize = '29')
plt.xlabel('$\sqrt{N}$' , color = 'black' , fontsize = '25' )
plt.ylabel('$d_{rms}$ (m)' , color = 'black' , fontsize = '25' )

plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)

plt.legend(loc = "best" , prop={'size':18} , shadow = True , edgecolor='black' , facecolor = 'wheat' )

plt.savefig( '7-34-a-app.pdf' )
```
