```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
coln = ['x', 'y', 't']
datos = pd.read_table('6-19-app.dat', sep='\s+', names=coln, header=None)

datos
```

```python
x = datos['x']
y = datos['y']
t = datos['t']

xx = np.array(x)
yy = np.array(y)
tt = np.array(t)
```

```python
import matplotlib.tri as tri

import matplotlib.cm as cm
cool = cm.get_cmap('rainbow')

plt.style.use('ggplot')

triang = tri.Triangulation(x, y)

fig, ax = plt.subplots(figsize=(16,9))
ax.set_aspect('equal')
tcf = ax.tricontourf(triang, t, 42, cmap = cool)
fig.colorbar(tcf , label = 'Temperatura en $C^{o}$' )
ax.tricontour(triang, t, colors='k' , levels = [12.5,25,37.5,50,62.5,75,87.5,95,100] , alpha = 0.75) 
ax.set_title('Temperatura en estado estacionario' , fontsize = '25')
ax.set_xlabel('$x$ (m)' , color = 'k' , fontsize = '20')
ax.set_ylabel('$y$ (m)' , color = 'k' , fontsize = '20')

plt.savefig('6-19-jup2-app.pdf')
```
