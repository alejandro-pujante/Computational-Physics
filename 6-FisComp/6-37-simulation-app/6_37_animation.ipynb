```python
# preparamos el grfico

fig, ax = plt.subplots(figsize=(10,10))
ax.set_xlim(( -5 , 505 ))
ax.set_ylim(( -0.17 , 0.17 ))
ax.set_xlabel('$x$ (cm)' , fontsize = '20')
ax.set_ylabel('$y$ (cm)' , fontsize = '20')
ax.set_title('Rope pulse with 2 diferent densities from the middle \n  $\\rho_{1}$ < $\\rho_{2}$ ', fontsize= '20')


x= np.linspace(0 , 500 , 1000)

[cuerda] = ax.plot( [] , [], '-' , color='darkred' , linewidth = '4.5')

[puntos]  = ax.plot(  [] , [] , '.' , markersize = '20' , color = 'k')

[linea] = ax.plot( [] , []  , '--' , color = 'gray' , alpha = 0.6)

plt.close();

def animate(k):
    global x 
   
    r = ([ x , datos.iloc[k] ])  

    cuerda.set_data( x , datos.iloc[k] )

    puntos.set_data( [0 , 500] , [0,0]  )

    linea.set_data( [250,250] , [-0.18,0.18]   )

    return ()

animation.FuncAnimation(fig, animate, init_func=lambda:[],frames=45*30, interval=300/30, blit=True)
```
